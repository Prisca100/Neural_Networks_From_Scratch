```python
import numpy as np
from deep_nn import *
from utils import train_test_split
```

```python
np.random.seed(1)
# Neural net architecture (number of layers in the network and number of nodes per layer including input and output nodes)
layer_dims = [12288, 10, 7,  1]

# 3 features by 100 samples
X = np.random.choice([0, 255], size=(12288, 259))


# 1 result for 100 samples
y = np.random.choice([0,1], size=(1, 259))

X_train, y_train, X_test, y_test = train_test_split(X, y, split_size=0.2)
X_train = X_train / 255
X_test = X_test / 255
parameters = initialize_parameters(layer_dims)
```

```python
# Computing the y_hat or predicted value
AL, caches = forward_pass(X_train, parameters)
print(caches)
```

```python
# cost for a single iteration
cost = compute_cost(AL, y_train)
print(cost)
```

```python
# Computing gradients using the chain rule
grads = backward_pass(AL, y_train, caches)
```

```python
# Getting new weights and biases
parameters = update_parameters(parameters, grads, learning_rate=0.15)
```

```python
# Putting it all together
params, costs, accuracies, lr = neural_net(X_train, y_train, layer_dims, learning_rate=0.05, num_iterations=1000, print_cost=True)
```

```python
plot_history(costs,accuracies, lr)
```

```python
predictions, accuracy = predict(X_test, y_test, params)
predictions
```
