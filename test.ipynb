```python
import numpy as np
from deep_nn import initialize_parameters,  forward_pass, compute_cost, backward_pass, update_parameters, neural_net, predict, plot_history
```

```python
np.random.seed(1)
# Neural net architecture (number of layers in the network and number of nodes per layer including input and output nodes)
layer_dims = [3, 20, 10, 10, 1]

# 3 features by 100 samples
X = np.random.randn(3, 100)

# 1 result for 100 samples
Y = np.random.choice([0,1], size=(1, 100))
parameters = initialize_parameters(layer_dims)
```

```python
# Computing the y_hat or predicted value
AL, caches = forward_pass(X, parameters)
print(f"AL: {AL}")
```

```python
# cost for a single iteration
cost = compute_cost(AL, Y)
print(cost)
```

```python
# Computing gradients using the chain rule
grads = backward_pass(AL, Y, caches)
```

```python
# Getting new weights and biases
parameters = update_parameters(parameters, grads, learning_rate=0.15)
```

```python
# Putting it all together
params, costs, accuracies, lr = neural_net(X, Y, layer_dims, learning_rate=0.1, num_iterations=2000, print_cost=True)
```

```python
plot_history(costs,accuracies, lr)
```

```python
predictions, accuracy = predict(X, Y, params)
predictions
```
